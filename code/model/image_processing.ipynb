```python
import cv2
import os

def resize_images_in_folder(folder_path, target_size):
    images = os.listdir(folder_path)

    for image_name in images:
        image_path = os.path.join(folder_path, image_name)
        image = cv2.imread(image_path)
        
        # Resize the image while maintaining the aspect ratio
        resized_image = cv2.resize(image, target_size)
        
        # Save the resized image back to the same location
        cv2.imwrite(image_path, resized_image)
```

```python
resize_shape = (150,150)
for dir in os.scandir('autism_image_data'):
    # Resize each valid subdir
    if dir.name[0] != '.':    
        resize_images_in_folder(f'autism_image_data/{dir.name}/Autistic', resize_shape)
        resize_images_in_folder(f'autism_image_data/{dir.name}/Non_Autistic', resize_shape)
```

```python
# Test that images were resized properly
display_images_from_folder('autism_image_data/valid/Autistic/', 5)
```
